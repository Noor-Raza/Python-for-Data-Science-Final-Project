### **Exercise 3:**

For each company, find the day with the highest Close price and display it alongside the price.

```python
from datetime import datetime

result = []  # List to store results
set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as the dataframe
    
    max_close_value = df['Close'].max()  # Extract the highest Close price
    max_close_row = df[datasets[dataset]['Close'] == max_close_value]  # Extract the row corresponding to the highest Close price
    max_close_date = max_close_row['Date'].iloc[0].date()  # Extract the date corresponding to the highest Close price
    
    formatted_max_close_date = datetime.strptime(str(max_close_date), '%Y-%m-%d').strftime('%d. %B %Y')  # Format the resulting date for improved readability

    result.append([dataset, f"{max_close_value:.2f}", formatted_max_close_date])  # Append the results to the list

    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nFor {dataset} the day with the highest Close price of {max_close_value:.2f} is {formatted_max_close_date}.')

result_df = pd.DataFrame(result, columns=['Dataset', 'Highest Close Price', 'Date'])  # Convert the result list to a pandas DataFrame

result_df
```

### **Exercise 4:**

Group the data by month and calculate the average Close price for each company. Plot these monthly averages for 3 companies and compare them. Justify the chart selection.

```python
all_avg_close_month = {}  # Initialize dictionary for all companies monthly average Close prices

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset].copy()  # Set df as copy of the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    avg_close_month = df.groupby(df['Month'])['Close'].mean()  # Calculate the average Close price by month

    all_avg_close_month[dataset] = avg_close_month  # Combine monthly averages into previously initialized dictionary

all_avg_close_month_df = pd.DataFrame(all_avg_close_month)  # Convert dictionary into DataFrame

print('Monthly Average Close Prices:')
round(all_avg_close_month_df, ndigits=2)  # Round result values to 2 decimals
```

```python
companies = ['DPZ', 'SBUX', 'YUM']  # Select three companies to plot charts

plt.figure(figsize=(15, 5))  # Set figure size

for company in companies:  # Iterate through the selected companies
    plt.plot(all_avg_close_month_df.index.to_timestamp(), all_avg_close_month_df[company], color=colors[company], label=company)  # Plot the average Close price by month as line graph
    plt.annotate(company, (pd.to_datetime('2024-10-01'), all_avg_close_month_df[company].iloc[-1]))  # Set dataset label at the end of the graph

company_names = ", ".join(companies)  # Extract the names out of the list of selected companies
plt.title(f'Comparison of Monthly Average Close Prices for {company_names}', fontweight='bold')  # Set graph title

plt.xlabel('Month')  # Set graph x axis label
plt.ylabel('Average Close Price')  # Set graph y axis label
plt.xlim(pd.to_datetime('2004-01-01'))  # Set graph start as last company data start
plt.grid(alpha=0.3)  # Add grid for better readability
plt.show()  # Display graph
```

#### **Justification of Chart Selection:**
We chose a line chart because it is the most effective way to compare the monthly average Close prices of the three companies over time.
- **Shows Trends Clearly:** A line chart allows us to visualize the trends in stock prices across months, highlighting the rises and falls for each company.
- **Ensures Fair Comparison:** We started the data from the point where all three companies had available data, ensuring that the comparison is meaningful and consistent across the timeline.
- **Represents Continuity:** Since the data is continuous over time, the lines provides a clear picture of how the average Close prices evolve month by month.
- **Improves Readability:** The gridlines and distinct colors for each company enhance readability, ensuring that the chart is easy to interpret.
- **Plotting from 2004:** The graph starts from 2004 because one of the selected companies has data available only from this year onwards. This ensures consistency in the timeline and avoids gaps in the plotted data.

### **Exercise 7:**

Create a plot showing the relationship between trading volume and the Close price for a selected company. Add insights about the pattern, and, justify your chart selection.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

company = 'DNUT'  # Select the company to plot the chart
df = datasets[company]  # Select the dataset for the company

# Create joint plot with scatter and regression line
jointplot = sns.jointplot(x='Volume', y='Close', data=df, kind="scatter", color=colors[company], height=9, ratio=4
                          ).plot_joint(sns.regplot, scatter=False, color='blue', line_kws={"linewidth": 2})

# Access the main plot in the jointplot
ax = jointplot.ax_joint

# Fit linear regression and calculate regression line
x = df['Volume']
y = df['Close']
coeffs = np.polyfit(x, y, 1)  # Fit a linear regression
reg_line = coeffs[0] * x + coeffs[1]  # Regression line

# Calculate standard deviation of residuals
std_dev = np.std(y - reg_line)

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Add regression line to the legend with correlation value
ax.plot(x, reg_line, color='blue', linewidth=2, label=f'Regression Line (Corr = {correlation:.2f})')

# Add 1 STD lines (upper and lower lines)
ax.plot(x, reg_line + std_dev, color='orange', linestyle='--', linewidth=1.5, label='+1 STD')  # Upper funnel
ax.plot(x, reg_line - std_dev, color='green', linestyle='--', linewidth=1.5, label='-1 STD')  # Lower funnel

# Use suptitle to add a title to the entire figure
jointplot.fig.suptitle(f"Joint Plot: Volume vs Close Price for {company}", fontweight='bold', y=1.02)

# Add axis labels and legend
ax.set_xlabel('Volume')
ax.set_ylabel('Close Price')
ax.legend()  # Add legend to the main plot
ax.grid(alpha=0.3)  # Add grid for better readability

# Adjust the figure layout to make it more landscape
jointplot.fig.set_size_inches(12, 6)  # Set the figure size to be more landscape

plt.show()  # Display graph
```

#### **Justification of Chart Selection:**
We chose a scatter plot embedded in a joint plot to visualize the relationship between trading volume and the Close price for the selected company for the followoing reasons:
- **Highlights the Relationship:** A scatter plot is the best choice for visualizing the correlation between two continuous variables (trading volume and Close price), making it easy to identify patterns or trends.
- **Density Insights:** The joint plot enhances the scatter plot by adding marginal histograms or density plots, providing additional insights into the distribution of both variables independently.
- **Identifies Outliers:** This chart type is ideal for spotting outliers that may indicate unusual trading activities or price behaviors.
#### **Understanding from the Chart Selection:**
- **Volume Concentration:** The marginal histogram for the trading volume shows a high concentration of data points around the lower volume range (below 1 million), indicating that most trading activity happens within this range.
- **Price Stability:** The Close price shows limited variability, with the majority of values clustering between $12 and $15. This suggests relatively stable pricing over the observed period, even with varying trading volumes.
- **Outlier Detection:** The scatter plot reveals some outliers where trading volume exceeds 3 million, and Close prices drop below $12. These outliers could indicate significant market events or anomalies.
- **Lack of Strong Trend:** The overall scatter pattern suggests no strong or consistent relationship between trading volume and Close price, implying that factors other than volume might influence the stock price.

### **Exercise 8:**

For each company, identify the month with the highest total trading volume and display the results in a summary table, showing the month and total volume.

```python
set_counter = 0  # Initialize set counter
all_max_volume_summary_table = {}  # Initialize dictionary for all company's highest total trading Volume

for dataset in datasets:  # Iterate through dictionary
    max_volume_summary_table = {}  # Initialize dictionary for each company's highest total trading Volume

    df = datasets[dataset].copy()  # Set a copy of df as the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    volume_month = df.groupby(df['Month'])['Volume'].sum()  # Calculate the total Volume by month

    max_volume_month = volume_month.idxmax()  # Extract the month corresponding to the highest Volume
    max_volume_value = volume_month.max()  # Extract the highest Volume

    formatted_max_volume_month = datetime.strptime(str(max_volume_month), '%Y-%m').strftime('%B %Y')  # Format the resulting date for improved readability
    
    max_volume_summary_table['Month'] = max_volume_month  # Enter month corresponding to the highest Volume for each company
    max_volume_summary_table['Volume'] = max_volume_value  # Enter the highest Volume for each company
    all_max_volume_summary_table[dataset] = max_volume_summary_table  # Combine values into previously initialized dictionary

    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nFor {dataset} the month with the highest total trading volume of {max_volume_value:.0f} is {formatted_max_volume_month}.')

all_max_volume_summary_table_df = pd.DataFrame(all_max_volume_summary_table)  # Convert dictionary into DataFrame

print('Summary Table of Highest Total Trading Volume by Company:')
round(all_max_volume_summary_table_df, ndigits=2)  # Round result values to 2 decimals
```

### **Exercise 9:**

Merge the datasets for all companies into a single dataset, one for each year. Print the structure of the combined dataset, ensuring proper alignment and handling of missing values.

```python
years = []  # Initialize list of years
merged_yearly_df = {}  # Initialize dictionary for the merged datasets for all years for all companies

for year in all_avg_close_year_df.index:  # Iterate through yearly dataframe indexes to extract all relevant years
    years.append(year)  # Store each year in the previously initialized list

for year in years:  # Iterate through all years
    yearly_df = {}  # Initialize dictionary for the merged datasets for each year for all companies

    for dataset in datasets:  # Iterate through dictionary of companies
        df = datasets[dataset].copy()  # Set df as copy of the dataframe
        
        df['Year'] = df['Date'].dt.year  # Add a new column for the year

        df.set_index(['Date'], inplace=True)  # Set Date as index

        yearly_df[dataset] = df[df['Year'] == year].drop(columns=['Year'])  # Filter data for the year, then drop the temporary year column and add to dictionary

    all_yearly_df = pd.concat(yearly_df.values(), keys=yearly_df.keys())  # Concatenate yearly dictionaries into DataFrame

    merged_yearly_df[year] = all_yearly_df  # Combine all yearly datasets into previously initialized dictionary

all_merged_yearly_df = pd.concat(merged_yearly_df.values(), keys=merged_yearly_df.keys())  # Concatenate merged dictionaries into DataFrame

all_merged_yearly_df
```

### **Exercise 10:**

For each company:
- Calculate the spread between the High and Low prices for each day
- Compute the average spread for each company and visualize the result in a chart. Justify the chart selection
- Interpret which companies exhibit the largest spreads and why this might happen

```python
set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Spread'] = df['High'] - df['Low']  # Add new column with the Spread between High and Low prices
    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe spread for {dataset} is the following:\n{df["Spread"]}')
```

```python
all_avg_spread = {}  # Initialize dictionary for all companies average Spread
bar_colors = []  # Initialize list of colors for each company

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Avg Spread'] = df['Spread'] / ((df['High'] + df['Low']) / 2)  # Add new column with the average Spread between the average of High and Low prices
    all_avg_spread[dataset] = df['Avg Spread'].mean()*100  # Combine average Spreads into previously initialized dictionary

    bar_colors.append(colors[dataset])  # Add color to the previously created list

plt.figure(figsize=(15, 5))  # Set figure size
plt.bar(all_avg_spread.keys(), all_avg_spread.values(), color=bar_colors)  # Plot the average Spreads as bar chart
plt.title(f'Comparison of Average Spreads by Company', fontweight='bold')  # Set chart title
plt.xlabel('Company')  # Set chart x axis label
plt.ylabel('Average Spread')  # Set chart y axis label
plt.grid(alpha=0.3)  # Add grid for better readability
plt.show()  # Display chart
```

```python
chart_colors = []  # Initialize list of colors for each company

for dataset in datasets:  # Iterate through dictionary
    chart_colors.append(colors[dataset])  # Add color to the previously created list

# Initialize dictionaries to hold monthly averages
monthly_avg_spreads = {}
last_5_years_monthly_avg_spreads = {}

# Loop through datasets to calculate monthly averages
for dataset in datasets:
    df = datasets[dataset].copy()

    # Calculate average spread
    df['Avg Spread'] = df['Spread'] / ((df['High'] + df['Low']) / 2)

    # Add a month-year column for grouping
    df['Month-Year'] = df['Date'].dt.to_period('M')

    # Calculate monthly average spreads
    monthly_avg_spreads[dataset] = df.groupby('Month-Year')['Avg Spread'].mean() * 100

    # Filter data for the last 5 years and calculate monthly averages
    last_5_years_data = df[df['Date'] >= pd.Timestamp.now() - pd.DateOffset(years=5)]
    last_5_years_monthly_avg_spreads[dataset] = last_5_years_data.groupby('Month-Year')['Avg Spread'].mean() * 100

# Combine all data into a single DataFrame for easier plotting
all_monthly_avg_spreads = pd.DataFrame(monthly_avg_spreads)
last_5_years_avg_spreads_df = pd.DataFrame(last_5_years_monthly_avg_spreads)

# Filter data to include only relevant periods (e.g., starting from 1980 if applicable)
filtered_all_monthly_avg_spreads = all_monthly_avg_spreads[all_monthly_avg_spreads.index >= '1980']
filtered_last_5_years_avg_spreads = last_5_years_avg_spreads_df[last_5_years_avg_spreads_df.index >= '1980']

# Plot the overall monthly average spread
filtered_all_monthly_avg_spreads.plot(color=chart_colors, figsize=(15, 5), alpha=0.7, linewidth=2)
plt.title('Monthly Average Spread by Company (Overall)', fontweight='bold', fontsize=14)
plt.xlabel('Month-Year')
plt.ylabel('Average Spread (%)')
plt.grid(alpha=0.5, linestyle='--', linewidth=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title='Company', fontsize=10)
plt.tight_layout()
plt.show()

# Plot the last 5 years' monthly average spread
filtered_last_5_years_avg_spreads.plot(color=chart_colors, figsize=(15, 5), alpha=0.7, linestyle='--', linewidth=2)
plt.title('Monthly Average Spread by Company (Last 5 Years)', fontweight='bold', fontsize=14)
plt.xlabel('Month-Year')
plt.ylabel('Average Spread (%)')
plt.grid(alpha=0.5, linestyle='--', linewidth=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title='Company', fontsize=10)
plt.tight_layout()
plt.show()
```
