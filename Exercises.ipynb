### Exercise 3:

For each company, find the day with the highest Close price and display it alongside the price.

#### Option 1

```python
from datetime import datetime

# List to store results
result = []

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as the dataframe
    
    max_close_value = df['Close'].max()  # Extract the highest Close price
    max_close_row = df[datasets[dataset]['Close'] == max_close_value]  # Extract the row corresponding to the highest Close price
    max_close_date = max_close_row['Date'].iloc[0].date()  # Extract the date corresponding to the highest Close price
    
    formatted_max_close_date = datetime.strptime(str(max_close_date), '%Y-%m-%d').strftime('%d. %B %Y')  # Format the resulting date for improved readability

    # Append the results to the list
    result.append([dataset, f"{max_close_value:.2f}", formatted_max_close_date])

# Convert the result list to a pandas DataFrame
result_df = pd.DataFrame(result, columns=['Dataset', 'Highest Close Price', 'Date'])

result_df
```

```python
import plotly.graph_objects as go

set_counter = 0  # Initialize counter for number of sets
all_avg_close_month = {}  # Initialize dictionary for all companies monthly average Close prices

# Loop through datasets and calculate monthly averages
for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset].copy()  # Set df as copy of the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    avg_close_month = df.groupby(df['Month'])['Close'].mean()  # Calculate the average Close price by month

    all_avg_close_month[dataset] = avg_close_month  # Combine monthly averages into previously initialized dictionary

# Convert dictionary into DataFrame
all_avg_close_month_df = pd.DataFrame(all_avg_close_month).reset_index()  # Reset index to include 'Month'

# Convert the 'Month' column from Period to string
all_avg_close_month_df['Month'] = all_avg_close_month_df['Month'].astype(str)

# Create the interactive table using Plotly
fig = go.Figure(data=[go.Table(
    header=dict(values=list(all_avg_close_month_df.columns),
                fill_color='paleturquoise',
                align='left',
                font=dict(size=12)),
    cells=dict(values=[all_avg_close_month_df[col].tolist() for col in all_avg_close_month_df.columns],
               fill_color='lavender',
               align='left',
               font=dict(size=10))
)])

# Show the table
fig.show()

```

### Exercise 8:

For each company, identify the month with the highest total trading volume and display the results in a summary table, showing the month and total volume.

#### Option 1

```python
all_max_volume_summary_table = {}  # Initialize dictionary for all company's highest total trading Volume

for dataset in datasets:  # Iterate through dictionary
    max_volume_summary_table = {}  # Initialize dictionary for each company's highest total trading Volume

    df = datasets[dataset].copy()  # Set a copy of df as the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    volume_month = df.groupby(df['Month'])['Volume'].sum()  # Calculate the total Volume by month

    max_volume_month = volume_month.idxmax()  # Extract the month corresponding to the highest Volume
    max_volume_value = volume_month.max()  # Extract the highest Volume
    
    max_volume_summary_table['Month'] = max_volume_month  # Enter month corresponding to the highest Volume for each company
    max_volume_summary_table['Volume'] = max_volume_value  # Enter the highest Volume for each company
    all_max_volume_summary_table[dataset] = max_volume_summary_table  # Combine values into previously initialized dictionary

all_max_volume_summary_table_df = pd.DataFrame(all_max_volume_summary_table)  # Convert dictionary into DataFrame

print('Summary Table of Highest Total Trading Volume by Company:')
round(all_max_volume_summary_table_df, ndigits=2)  # Round result values to 2 decimals
```

### Exercise 9:

Merge the datasets for all companies into a single dataset, one for each year. Print the structure of the combined dataset, ensuring proper alignment and handling of missing values.

#### Option 1

```python
years = []  # Initialize list of years
merged_yearly_df = {}  # Initialize dictionary for the merged datasets for all years for all companies

for year in all_avg_close_year_df.index:  # Iterate through yearly dataframe indexes to extract all relevant years
    years.append(year)  # Store each year in the previously initialized list

for year in years:  # Iterate through all years
    yearly_df = {}  # Initialize dictionary for the merged datasets for each year for all companies

    for dataset in datasets:  # Iterate through dictionary of companies
        df = datasets[dataset].copy()  # Set df as copy of the dataframe

        df['Year'] = df['Date'].dt.year  # Add a new column for the year

        yearly_df[dataset] = df[df['Year'] == year].drop(columns=['Year'])  # Filter data for the year, then drop the temporary year column and add to dictionary

    all_yearly_df = pd.concat(yearly_df.values(), keys=yearly_df.keys())  # Concatenate yearly dictionaries into DataFrame

    merged_yearly_df[year] = all_yearly_df.reset_index(level=1, drop=True).reset_index()  # Combine all yearly datasets into previously initialized dictionary and reset indexes

all_merged_yearly_df = pd.concat(merged_yearly_df.values(), keys=merged_yearly_df.keys())  # Concatenate merged dictionaries into DataFrame

all_merged_yearly_df
```

```python
rows, columns = all_merged_yearly_df.shape  # Extract the row and column of merged dataset
column_names_types = all_merged_yearly_df.dtypes  # Extract the column names and data types of merged dataset

print(f'The merged datasets have {rows} rows and {columns} columns.\nThe column names and their data types are the following:')

for column_name, column_type in column_names_types.items():  # Iterate through all columns and their types
    print(f' - {column_name}: {column_type}')

print('Additionally, there are multiple layered indexes for the Year, Company and initial row.')
```

- Compute the average spread for each company and visualize the result in a chart. Justify the chart selection
- Interpret which companies exhibit the largest spreads and why this might happen

#### Option 1

```python
set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Spread'] = df['High'] - df['Low']  # Add new column with the Spread between High and Low prices
    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe spread for {dataset} is the following:\n{df["Spread"]}')
```

```python
all_avg_spread = {}  # Initialize dictionary for all companies average Spread
bar_colors = []  # Initialize list of colors for each company

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Avg Spread'] = df['Spread'] / ((df['High'] + df['Low']) / 2)  # Add new column with the average Spread between the average of High and Low prices
    all_avg_spread[dataset] = df['Avg Spread'].mean()*100  # Combine average Spreads into previously initialized dictionary

    bar_colors.append(colors[dataset])  # Add color to the previously created list

plt.figure(figsize=(15, 5))  # Set figure size
plt.bar(all_avg_spread.keys(), all_avg_spread.values(), color=bar_colors)  # Plot the average Spreads as bar chart
plt.title(f'Comparison of Average Spreads by Company')  # Set chart title
plt.xlabel('Company')  # Set chart x axis label
plt.ylabel('Average Spread')  # Set chart y axis label
plt.grid(alpha=0.3)  # Add grid for better readability
plt.show()  # Display chart
```
